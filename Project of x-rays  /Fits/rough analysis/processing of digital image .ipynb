```python
import matplotlib.pyplot as plt
import os
import numpy as np
import os.path
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
for i in range(-10,0):
    os.system("sed -i -- 's/+0.*I/ /g' dynamicaltheory/dynamicaltdata/rod=%s.txt"%i)
for j in range(1,15):
    theory=np.loadtxt('dynamicaltheory/dynamicaltdata/rod=%s.txt'%(-j+0))
    z=theory[:,0]
    inti=300000000*theory[:,1]
    v=np.loadtxt('../../processing/data_0.2deg/average/l=%s.dat'%j)
    a=v[:,0]
    b=v[:,1]
    plt.title('m=%s'%j)
    plt.plot(10*a,b/np.max(b),label='Exp')
    plt.plot(z,10*smooth(inti,1)/np.max(smooth(inti,1)),label='smooth',color='black')
    plt.yscale('log')
    plt.legend()
    plt.xlim(-0.5,1.7)
    print(j)
    plt.gcf().savefig('m=%s.png'%j,dpi=700)
    plt.show()
```
