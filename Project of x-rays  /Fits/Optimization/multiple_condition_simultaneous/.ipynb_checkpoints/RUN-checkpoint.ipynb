```python
import cma
import os
import numpy as np
import func 
import time
import shutil
import matplotlib.pylab as plt
```

This part is for determining input parameters

### Structural input

```python
code='new.wls' #name of the mathematica code
d=141 #periodic length of grating 140
fd=4 #the filling factor out of 10
fbd=6#filling factor of bottom layer out of 10
ftd=4.5 #filling factor of top layer out of 10
t=31 #grating thickness middle part
tc=10 #thickness of other parts
```

### Experimental data path

```python
exp='exp/average/'
rods_number=[3,4,5]
```

### Cma-es package input

```python
dimension=6 #number of structural parameters to be predicted.
tol=10e-6
X=[d,fd,fbd,ftd,t,tc]#series of initial guess 
popsize=5 #number of different solutions each time package provides, and need to be evaluated
std=4.45
```

### Output files

```python
output=open('output.txt','w')
log=open('log.txt','w')
product=open('product.txt','w')
cnt=0
```

### Code criteria 

```python
threshold=3 #minium which result should be in this span to have an acceptable result
sleep_time=30 #in seconds. This is the time for "popsize" number of simulation to be completed. 
cnt=0
```

### Function for fitting

```python
Ubound=np.array([150,7.25,7.25,7.25,50,50])
Lbound=np.array([130,2.75,2.75,2.75,10,10])
es = cma.CMAEvolutionStrategy( X, std,{'popsize':popsize,'tolfun':tol,'bounds':[Lbound,Ubound]})
```

```python
while not es.stop():
    fit=np.zeros((1,popsize))
    solutions=es.ask()     
    print('start the game')
    for i in range(popsize):
        print(solutions[i])
        print(func.seq(np.array(solutions[i]).tolist()))
        print('func.simulation6(%s,%s)'%(code,func.seq(np.array(solutions[i]).tolist())))
        v=np.array(solutions[i]).tolist()
        s=func.simulation6(code,v[0],v[1],v[2],v[3],v[4],v[5]) 
       # shutil.move('sample.txt', 'sample7.txt')
       # os.system('mv C:\Users\hrahmani\Desktop\Optimization\Process_of_fitting\Main\test\multiple_condition_simultaneous_three_variables\sample.txt C:\Users\hrahmani\Desktop\Optimization\Process_of_fitting\Main\test\multiple_condition_simultaneous_three_variables\sample7.txt')
       # n=func.simulation4(code,v[0],v[1],v[2],v[3],v[4],v[5])
      #  shutil.move('sample.txt', 'sample4.txt')
        print('v0=%s v1=%s v2=%s v3=%s v4=%s v5=%s'%(v[0],v[1],v[2],v[3],v[4],v[5]))
        time.sleep(sleep_time)
        print('sleep time is over')
        
        for j in range(-15,0):
            fff=np.loadtxt('dyt/rod=%s.txt'%(j))
            maxi=np.max(fff[:,1])
            dta=open('dyt/rodn=%s.txt'%j,'w')
            for k in range(len(fff)):
                if fff[k,1]>0 and fff[k,0]<0.8:
                    dta.write('%s %s\n'%(fff[k,0],fff[k,1]/maxi))
            dta.close()
        fitn_matrix=[]
        for m in rods_number:
            sim_data=('dyt/rodn=-%s.txt'%m)
            exp_data=('exp/average/l=%s.dat'%m)
            fit_number=func.fitness(sim_data,exp_data)
            print('fitness number for rod %s is %s'%(m,fit_number))
            fitn_matrix.append(fit_number)
            #visualization
            thery=np.loadtxt('dyt/rodn=-%s.txt'%m)
            exyp=np.loadtxt('exp/average/l=%s.dat'%m)
            plt.plot(thery[:,0],thery[:,1],label='theory')
            plt.plot(exyp[:,0],exyp[:,1],label='exyp')
            plt.legend()
            plt.yscale('log')
            plt.show()           
        fitness_number=np.mean(fitn_matrix)
        print('fitness is %s'%fitness_number)
        fit[0][i]=fitness_number
        log.write('%s simulation number%s\n'%(cnt,i))
        print('%s simulation number%s\n'%(cnt,i))
    fit=np.array(fit).tolist()[0]
    output.write('%s %s %s \n'%(cnt,solutions,fit)) 
    print('%s %s %s \n'%(cnt,solutions,fit))
    es.tell(solutions,fit)
    ftn=es.result[1]
    newsol=es.result[0]
    cnt+=1
    output.write('new solution is %s with %s as fitness criteria\n\n #### \n'%(newsol,ftn))
    print('new solution is %s with %s as fitness criteria\n\n##########\n'%(newsol,ftn))
    product.write('%s %s\n'%(cnt,ftn)) 
    
    if (np.abs(ftn)<threshold):
        break
        output.write('last line shows the solution')
output.close()
log.close()
product.close()
    
    
```
