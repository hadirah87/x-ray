```python
import matplotlib.pyplot as plt
import os
import numpy as np
import os.path
from scipy.signal import find_peaks
```

```python
directory='../Data/dat/datswep/0.2deg/' #path to the digital image data
lst=os.listdir(directory)
direct=os.path.abspath('../Data/dat/datswep/0.2deg/')
lst.sort()
print(lst)
```

```python
s1=np.loadtxt('../Data/dat/datswep/0.2deg/SNS_C72_16.1keV0.2deg_sweep20deg_000001_SAXS.dat')
peaks, _ = find_peaks(s1[531,:],threshold=None,distance=23)
print(peaks)
#rods=peaks[int(len(peaks)/2):]
print(len(peaks))
rods=peaks[int(len(peaks)/2)+3:]
print(rods)
pixel_size=172e-6
total_pixel_y_dir=1043
center_of_y_h=0.059#center of digital image is pixel number (x,y)=(556,1043-614)*172e-6=(0.095632,0.073788)
print(len(peaks))
print(len(rods))
```

```python
for item in lst:
    if os.path.isdir('data_0.2deg/folder%s'%(item[:-4])):
        pass
    else:
        os.mkdir('data_0.2deg/folder%s'%(item[:-4]))
    ary=np.loadtxt('%s%s'%(directory,item))
    qy=ary[601,:]    
    for i in range(len(rods)):
        qz=ary[:,rods[i]]
 #       print(qz)
        np.savetxt('data_0.2deg/folder%s/%sm=%s.dat'%(item[:-4],item[:-4],i),np.c_[list(range(len(qz),0,-1)),qz],delimiter=' ')
        
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
for i in range(-15,0):
    os.system("sed -i -- 's/+0.*I/ /g' ../dynamical_theory_data/dynamicaltdatanew/rod=%s.txt"%i)
for j in range(1,15):
#    os.system("sed -i -- 's/+0.*I/ /g' ../Data_of_Dynamicaltheory/Data_of_specular/rod=%s.txt"%j)
    theory=np.loadtxt('../dynamical_theory_data/dynamicaltdatanew/rod=%s.txt'%(-j+0))
    z=theory[:,0]
    inti=300000000*theory[:,1]
    v=np.loadtxt('data_0.2deg/average/l=%s.dat'%j)
    a=v[:,0]
    b=v[:,1]
    plt.title('m=%s'%j)
    plt.plot(10*a,b/np.max(b),label='Exp')
    plt.plot(z,10*smooth(inti,1)/np.max(smooth(inti,1)),label='smooth',color='black')
#    plt.plot(z,inti,label='DT')
    plt.yscale('log')
    plt.legend()
    plt.xlim(-0.5,1.7)
    print(j)
    plt.gcf().savefig('m=%s.png'%j,dpi=700)
    plt.show()
```
