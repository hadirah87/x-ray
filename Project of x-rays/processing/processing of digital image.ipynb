# Tutorial for scattered x-ray image processing 

This code is a brief summary of how .dat images can be analyzed and visualized to extract information about bragg rods

### Importing necessary libraries

```python
import matplotlib.pyplot as plt
import os
import numpy as np
import os.path
from scipy.signal import find_peaks
```

### Creating directory to store neccessary data

```python
directory='../Data/dat/datswep/0.2deg/' #path to the digital image data, make directory named 0.2deg in datswep to store data
lst=os.listdir(directory)
direct=os.path.abspath('../Data/dat/datswep/0.2deg/')
lst.sort()
print(lst)
```

### Finding bragg rods on .dat image

```python
s1=np.loadtxt('../Data/dat/datswep/0.2deg/SNS_C72_16.1keV0.2deg_sweep20deg_000001_SAXS.dat')
peaks, _ = find_peaks(s1[531,:],threshold=None,distance=23)
print(peaks)
#rods=peaks[int(len(peaks)/2):]
print(len(peaks))
rods=peaks[int(len(peaks)/2)+3:] #only using the most intense bragg rods by looking at peaks in image
print(rods)
pixel_size=172e-6
total_pixel_y_dir=1043
center_of_y_h=0.059 #center of digital image is pixel number (x,y)=(556,1043-700)*172e-6=(0.095632,0.073788)
print(len(rods))
```

### Making directory 'data_0.2deg' to hold data

```python
for item in lst:
    if os.path.isdir('data_0.2deg/folder%s'%(item[:-4])): 
        pass
    else:
        os.mkdir('data_0.2deg/folder%s'%(item[:-4])) # making directory for data
    ary=np.loadtxt('%s%s'%(directory,item))
    qy=ary[601,:]    
    for i in range(len(rods)):
        qz=ary[:,rods[i]]
        np.savetxt('data_0.2deg/folder%s/%sm=%s.dat'%(item[:-4],item[:-4],i+1),np.c_[list(range(len(qz),0,-1)),qz],delimiter=' ')
        #np.savetxt stored the data in the new directory 'data_0.2deg'
```

### Visualizing data

```python
for item in lst[1:10]: # this cell plots intensity on a log scale vs the y axis of image for each bragg rod
    for i in range(1,10):
        fff=np.loadtxt('data_0.2deg/folder%s/%sm=%s.dat'%(item[:-4],item[:-4],i))
        print(item,i)
        x=fff[:,0]
        y=fff[:,1]
        #plt.plot(x,y)
        plt.plot((x*pixel_size-center_of_y_h)+0.0095,y)
        plt.yscale('log')
        #plt.xlim(0,0.125)
        plt.show()
```

### Getting rid of spikes of 0s in data ( not actaul data points and will mess up further analysis) 

```python
for i in range(1,15):
    newdata=open('data_0.2deg/average/final_average/m=%s.dat' %(i),'w')
    fff=np.loadtxt('data_0.2deg/average/m=%s.dat'%(i))
#        print(item,i)
    x=fff[:,0]
    y=fff[:,1]
    ymax=np.max(y)
    for j in range(len(y)):
        if y[j]<=0 or x[j]<0.005 or x[j]>0.087:
            pass
        else:
            newdata.write('%s %s\n'%(x[j]*10,np.log10(y[j]/ymax)))   
newdata.close()
```
